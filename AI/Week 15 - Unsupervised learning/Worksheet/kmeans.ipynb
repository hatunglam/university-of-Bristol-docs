```python
import numpy as np
import matplotlib.pyplot as plt
from kmeans import KMeans
```

```python
random_point = np.random.randint(0, 100, (100, 2))

kmeans = KMeans(k = 3)
labels = kmeans.fit(random_point)

plt.scatter(random_point[:, 0], random_point[:, 1], c= labels)
plt.scatter(kmeans.centroids[:, 0], kmeans.centroids[:,1], c= range(len(kmeans.centroids)),
            marker = '*', s= 200)
plt.show()
```
